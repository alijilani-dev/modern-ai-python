```python
########################################################################################################
# ------------>                 0       1       2       3       Forward Order
student_names : list[str] = ["Rehan", "Musa", "Imran", "Qasim"] # Sequenced, Ordered, Indexed Data Type.
# <------------                -4      -3      -2      -1       Reverse (-ve) Order (-ve step)
########################################################################################################
# list[start: end: +ve step]    starting index is 'last item:3' by default, ending index 'first index:0' by default and is excluded.
# list[start: end: -ve step]    starting index is 'last item:3' by default, ending index 'first index:0' by default and is excluded.
###########################################################################
student_names # student_names[:], student_names[::] All mean the same thing
###########################################################################
```

```python
# Indexing : Value Retrieval
print(student_names)
print(student_names[1], "Forward/+ve Indexing")   # Forward/ +ve Indexing
print(student_names[-1], "Reverse/-ve Indexing")  # Backward/ Reverse -ve Indexing

# Indexing : Value Assignment
student_names[1] = "Junaid"         # "Musa" Replaced by "Junaid"
print(student_names)
############################################################################################
# Both Forward and Reverse Orders work: Every List Item has a positive and a negative index
############################################################################################
print(student_names[1], "Forward Indexing: 1")   # Forward Indexing
print(student_names[-1], "Backward/ Reverse Indexing: -1") # Backward/ Reverse Indexing

# Slicing
# list[start: end: step]    starting index is 0 by default, ending index is 'last index' by default and is excluded, step is 0 by default.
print(student_names[1:3]," -> 1 & 3")
print(student_names[0:3:], " -> 0 & 3")
print(student_names[0:-2:], " -> 0 & -2")
print(student_names[0:-1:], " -> 0 & -1")
print(student_names[0:3:1], " -> 0 & 3 step 1")
########################################################################################################
# ------------>                 0       1       2       3       Forward Order
student_names : list[str] = ["Rehan", "Musa", "Imran", "Qasim"] # Sequenced, Ordered, Indexed Data Type.
# <------------                -4      -3      -2      -1       Reverse (-ve) Order (-ve step)
########################################################################################################
# Both Forward and Reverse Orders work: Every List Item has a positive and a negative index
###########################################################################################
print(student_names[::-1], " -> def & def step -1; starting index is 'last item:3' by default, ending index 'first index:0' by default and is excluded.")
print(student_names[3:1:-1], " -> 3 & 1 step -1   (Positive Indexes)")             # +ve start, +ve end     ['Qasim', 'Imran']
print(student_names[-1:-3:-1], " -> -1 & -3 step -1 (Negative Indexes)")           # -ve start, -ve end     ['Qasim', 'Imran']

###############
# Mixed Indexes
###############
print(student_names[-2:0:-1], " -> -2 & 0 step -1 (Mixed Indexes)")                # -ve start, +ve end)    ['Imran', 'Junaid']
print(student_names[-2:-4:-1], " -> -2 & -4 step -1 (Negative Indexes)")           # -ve start, -ve end)    ['Imran', 'Junaid']

#################################
# Assignment using range/ Slicing
#################################
student_names
student_names[0:2:] = "Qasim"
student_names         # Wierd
```
