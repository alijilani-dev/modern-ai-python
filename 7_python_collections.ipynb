```python
########################################################################################################
# ------------>                 0       1       2       3       Forward Order
student_names : list[str] = ["Rehan", "Musa", "Imran", "Qasim"] # Sequenced, Ordered, Indexed Data Type.
# <------------                -4      -3      -2      -1       Reverse (-ve) Order (-ve step)
########################################################################################################
# list[start: end: +ve step]    starting index is 'last item:3' by default, ending index 'first index:0' by default and is excluded.
# list[start: end: -ve step]    starting index is 'last item:3' by default, ending index 'first index:0' by default and is excluded.
###########################################################################
student_names # student_names[:], student_names[::] All mean the same thing
###########################################################################
# Indexing : Value Retrieval
print(student_names)
print(student_names[1], "Forward/+ve Indexing")   # Forward/ +ve Indexing
print(student_names[-1], "Reverse/-ve Indexing")  # Backward/ Reverse -ve Indexing

# Indexing : Value Assignment
student_names[1] = "Junaid"         # "Musa" Replaced by "Junaid"
print(student_names)
############################################################################################
# Both Forward and Reverse Orders work: Every List Item has a positive and a negative index
############################################################################################
print(student_names[1], "Forward Indexing: 1")   # Forward Indexing
print(student_names[-1], "Backward/ Reverse Indexing: -1") # Backward/ Reverse Indexing

# Slicing
# list[start: end: step]    starting index is 0 by default, ending index is 'last index' by default and is excluded, step is 0 by default.
print(student_names[1:3]," -> 1 & 3")
print(student_names[0:3:], " -> 0 & 3")
print(student_names[0:-2:], " -> 0 & -2")
print(student_names[0:-1:], " -> 0 & -1")
print(student_names[0:3:1], " -> 0 & 3 step 1")
########################################################################################################
# ------------>                 0       1       2       3       Forward Order                  -------->
student_names : list[str] = ["Rehan", "Musa", "Imran", "Qasim"] # Sequenced, Ordered, Indexed Data Type.
# <------------                -4      -3      -2      -1       Reverse (-ve) Order (-ve step) <--------
########################################################################################################
# Both Forward and Reverse Orders work: Every List Item has a positive and a negative index
###########################################################################################
print(student_names[::-1], " -> def & def step -1; starting index is 'last item:3' by default, ending index 'first index:0' by default and is excluded.")
print(student_names[3:1:-1], " -> 3 & 1 step -1   (Positive Indexes)")             # +ve start, +ve end     ['Qasim', 'Imran']
print(student_names[-1:-3:-1], " -> -1 & -3 step -1 (Negative Indexes)")           # -ve start, -ve end     ['Qasim', 'Imran']

###############
# Mixed Indexes
###############
print(student_names[-2:0:-1], " -> -2 & 0 step -1 (Mixed Indexes)")                # -ve start, +ve end)    ['Imran', 'Junaid']
print(student_names[-2:-4:-1], " -> -2 & -4 step -1 (Negative Indexes)")           # -ve start, -ve end)    ['Imran', 'Junaid']

#################################
# Assignment using range/ Slicing
#################################
student_names
####################################################################################
# To replace first two elements as a single value, we'd need to wrap them as a list.
####################################################################################
section_changed : list[str] = ["Shahzad","Amir"]
student_names[0:2:] = section_changed
student_names

###################################################################################################################################
# slice assignment unpacks iterables (string in python is an iterable, but of characters that gets unpacked while slice assignment)
###################################################################################################################################
student_names
student_names[0:2:] = "Qasim"
student_names         # Wierd : Slicing will not be used to update list elements. NumPy Array will be used subsequently.
```

```python
############################################################################################################################################################################
# Explicit Shallow Copy Method applied on list of immutable items, (I Feel Big Stones Turn Firm Under My Feet :  Int, Float, Bool, String, PENDING:Tuple, PENDING:Frozenset)
############################################################################################################################################################################
from typing import Any
student_info : list[Any] = [10, 20, 3.5, 4.0, True, True, "Up", "Down"]
student_info                   # List of immutable items only
student_info1 = student_info  # Shallow copy method (shared memory reference between both objects)
print(student_info)
print(student_info1)
print("---Testing effect of Shallow copy method on Integer immutable item - proves to be Shallow copy")
student_info1[0] = 30          # Only new variable (holding memory references of main variable) is changed
print(student_info)            # Shallow copy proved - Main variable is also changed
print(student_info1)
print("---Testing effect of Shallow copy method on Float immutable item - proves to be Shallow copy")
student_info1[2] = 3.7         # Only new variable (holding memory references of main variable) is changed
print(student_info)            # Shallow copy proved - Main variable is also changed
print(student_info1)
print("---Testing effect of Shallow copy method on Boolean immutable item - proves to be Shallow copy")
student_info1[4] = False       # Only new variable (holding memory references of main variable) is changed
print(student_info)            # Shallow copy proved - Main variable is also changed
print(student_info1)
print("---Testing effect of Shallow copy method on String immutable item - proves to be Shallow copy")
student_info1[6] = "Go"          # Only new variable (holding memory references of main variable) is changed
print(student_info)            # Shallow copy proved - Main variable is also changed
print(student_info1)
```

```python
##################################################################################################################################################################
# With Copy Function, Deep copy (big stones) occurs for immutable items in the list, and shallow copy incase of mutable items (little ducks swim in shallow water)
##################################################################################################################################################################
#student_names.copy()

#####################################################################
# Shallow copy happens on immutable and mutable elements in our list.
###############################################################################################################################
data : list[str | list] = ["Rehan", "Adnan", [1,2,3]]   # our data is a mix of immutable (strings) and mutable (list) elements.

data1 = data            # Shallow Copy method applied on mix of mutable and immutable

print("Data: ",data)
print("Data1: ",data1)

print("---Testing effect of Shallow copy method on immutable item (string) - proves to be Shallow copy")
data1[0] = "Muhammad Rehan"

print("Data: ",data)
print("Data1: ",data1)

print("---Testing effect of Shallow copy method on mutable item (list)- proves also to be Shallow copy")
data1[2][0] = 10

print("Data: ",data)
print("Data1: ",data1)
```

```python
#############################################################################################################################################################################
# Shallow Copy Function: kind of Deep copy (big stones) occurs for immutable items in the list, and shallow copy incase of mutable items (little ducks swim in shallow water)
#############################################################################################################################################################################
#student_names.copy()

#######################################################################################
# Shallow Copy Function/ Method applied on immutable and mutable elements in your list.
###############################################################################################################################
data : list[str | list] = ["Rehan", "Adnan", [1,2,3]]   # our data is a mix of immutable (strings) and mutable (list) elements.

data1 = data.copy()            # Shallow Copy method applied on mix of mutable and immutable

print("Data: ",data)
print("Data1: ",data1)

print("---Testing effect of Shallow copy method on immutable items - proves to be a Deep copy")
data1[0] = "Muhammad Rehan"

print("Data: ",data)
print("Data1: ",data1)

print("---Testing effect of Shallow copy method on mutable items - proves to be Shallow copy")
data1[2][0] = 10

print("Data: ",data)
print("Data1: ",data1)
```

```python
#############################################################################################################
# List Functions/ Methods: 'count', 'extend', 'append', 'index', 'insert', 'pop', 'remove', 'reverse', 'sort'
#############################################################################################################

names : list[str] = ["Qasim", "Qasim", "Rehan", "Ameen"]
print(names.count("Qasim"), "\n")
names.count("qasim")      # Case Sensitive


names1 : list[str] = ['Qasim', 'Rehan']
names2 : list[str] = ['Asif', 'Kashif']

print(names1 + names2)    # Inline addition operation
print('-----Extend Function-----')
names1.extend(names2)     # extend method adds as items to the same list.
print('Names1 Extended: ',names1)
names1.append(names2)     # append method adds the full list as a separate item to the end of first list.
print('Names1 Appended: ',names1)
names1.index('Qasim')     # index method returns the index of first occurrance of the item in list.
print('Occurrance Index: ',names.index('Rehan'))

#                       0        1        2        3
names : list[str] = ["Qasim", "Rehan", "Ameen", "Qasim"]
print('Occurrance Index Starting: ', names.index('Qasim',1))
names.insert(1, 'Ahmed')  # Item inserted at index 1.
print('Insert Item at Index 1:', names)

#                       0        1        2        3
names : list[str] = ["Qasim", "Rehan", "Ameen", "Qasim"]
names.insert(-2, 'Ahmed')  # Item inserted at index -2.
print('Insert Item at Index -2:', names)

data : list[int] = [2,1,3,7,8]
print('Data: ', data)
data.sort()               # sort data ascending by default.
print('Sorted Data: ', data)

data : list[int] = [2,1,3,7,8]
data.sort(reverse=True)   # reverse sort data (descending order)
print('Reverse Sort: ', data)

data : list[int] = [2,1,3,7,8]
data.reverse()            # Not sorted.
print('Reverse Data: ', data)

data : list[int] = [2,1,3,7,8]
data.sort()               # sort data in ascending order
data.reverse()            # reverse of ascending => descending.
print('Descending Sort: ', data)
```

```python
###############
# Popping-Cart
###############

to_buy_items: list[str]; items_bought : list[str] = []
while (b := input("Add item (or 'done' to stop): ")) != "done": to_buy_items.append(b)

print(to_buy_items)

while (c := input("Bought item (or 'done' to stop): ")) != "done" : items_bought.append(to_buy_items.pop(to_buy_items.index(c)))

print("To Buy: ",to_buy_items)
print("Bought: ",items_bought)

# Handle exceptions if Nothing to buy, Bought item is not in buy items.
```

```python
#######################
# Popping-Cart : Remove
#######################

to_buy_items : list[str] = []; items_bought : list[str] = []
while (b := input("Add item (or 'done' to stop): ")) != "done": to_buy_items.append(b)

print(to_buy_items)

while (c := input("Bought item (or 'done' to stop): ")) != "done" : to_buy_items.remove(c) # remove doesn't return item to append.

print("To Buy: ",to_buy_items)
print("Bought: remove doesn't return item to append..")
```
