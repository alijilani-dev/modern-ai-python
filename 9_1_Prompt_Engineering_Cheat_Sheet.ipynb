🚀 Prompt Engineering – Numerical Concepts (2025 Guide) 🚀

Here’s a cheat sheet of numbers & settings every AI engineer should know 👇

✨ LLM Output Settings

Temperature
• 0 → Deterministic (math, CoT).
• 0.1 → Low creativity (code, parsing).
• 0.2 → Balanced start (general use).
• 0.7 → Creative writing.
• 0.9 → Poetry/brainstorming.
• 1.0 → Max randomness.

Top-K
• 1 → Greedy decoding.
• 20 → Conservative.
• 30 → Balanced.
• 40 → Creative.

Top-P (Nucleus Sampling)
• 0.9 → Conservative.
• 0.95 → Balanced.
• 0.99 → Creative.

Token Limits
• 5 → Quick classification.
• 250 → Parsing tasks.
• 1024 → Role/system prompts.
• 1M → GPT-4.1 context window.

✨ Techniques & Best Practices

Few-shot: 3–6 examples.

Step-back: Ask model meta-questions.

Self-consistency: Multiple outputs → majority answer.

APE: Auto-generate + score prompts.

Context: 3–6 sentences (~100–150 words).

✨ Image Prompts (Numbers to Know)

📸 Lenses: 85mm (portrait), 35mm (wide), 50mm (normal).

🔦 Aperture: f/1.4 (bokeh) → f/11 (sharp).

🎨 Lighting: 2700K warm → 7000K+ cool.

✨ Video Prompts

⏱ Duration: 8 sec clips (best performance).

🎥 Styles: 35mm classic, 16mm doc, Super 8 vintage.

📐 Ratios: 16:9, 9:16, 1:1.

✨ MoE Models (2025)

GPT-5 → ~2T params, sparse experts.

DeepSeek-V3 → 671B (37B active).

Mixtral 8x7B → 8 experts/layer.

Only 10–20% experts active per token.

⚡ In short → Numbers matter. Master them, and your prompts perform at pro level.
